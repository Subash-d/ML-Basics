```python
import pandas as pd
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Experience Years'], df['Salary'])
plt.title('Salary based on EXP')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
X = df[['Experience Years']]
y = df['Salary']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_test_pred = model.predict(X_test)
y_test_pred 
```

```python
print(model.coef_)
print(model.intercept_)
# y = m x + c (co - efficient)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
R2_score = r2_score(y_test, y_test_pred)
MSE = mean_squared_error(y_test, y_test_pred)
MAE = mean_absolute_error(y_test, y_test_pred)
print('R2_score: ', R2_score)
print('Mean Squared Error: ', MSE)
print('Mean Absolute Error: ', MAE)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))

experience_test = X_test.squeeze()
experience_train = X_train.squeeze()

# Plot actual salary as a straight datapoints
sns.scatterplot(x=experience_test, y=y_test, label='Actual Salary', color='blue')

# Plot predicted salary as line
sns.lineplot(x=experience_test, y=y_test_pred, label='Predicted Salary', color='green')

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Actual vs Predicted Salary")
plt.legend()
```
