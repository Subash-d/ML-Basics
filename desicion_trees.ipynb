```python
import pandas as pd
df = pd.read_csv('Datasets/Comedy_show.csv')
df.head()
```

```python
Nationality_dummies = pd.get_dummies(df.Nationality)
Go_dummies = pd.get_dummies(df.Go)
```

```python
df = df.drop(['Nationality','Go'], axis=1)
df
```

```python
df = pd.concat([df,Nationality_dummies,Go_dummies], axis= 1)
df
```

```python
X = df.iloc[:, :-1]  
y = df.iloc[:, 7]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
dtc = DecisionTreeClassifier()
dtc.fit(X,y)
```

```python
tree.plot_tree(dtc)
```

```python
print(dtc.predict([[40, 10, 7, 1, True, True, False]]))
```
