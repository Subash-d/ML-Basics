```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = df[['Experience Years']]
y = df['Salary']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 20)
reg = LinearRegression()
fitted = reg.fit(X_train,y_train)
fitted
```

```python
fitted.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
y_predicted = reg.predict(X_test)
r2_score(y_predicted, y_test)
```

```python
mse = mean_squared_error(y_predicted, y_test)
mse
```

```python
mae = mean_absolute_error(y_predicted, y_test)
mae
```
