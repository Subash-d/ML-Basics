```python
#GRADIENT DESCENT

import numpy as np
import matplotlib.pyplot as plt

# Sample training data (you can change this)
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])

# Initialize parameters
m = 0     # slope
c = 0     # intercept

# Hyperparameters
learning_rate = 0.001
epochs = 4000
n = len(X)

# Training loop
for i in range(epochs):
    y_pred = m * X + c
    error = y - y_pred

    # Gradients
    dm = (-2/n) * sum(X * error)
    dc = (-2/n) * sum(error)

    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc

    mse = (error ** 2).mean()
    print(f"Epoch {i}: m = {m:.4f}, c = {c:.4f}, MSE = {mse:.4f}")

# Final line
print(f"\nFinal Equation: y = {m:.2f}x + {c:.2f}")

# Plot
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, m*X + c, color='red', label='Predicted Line')
plt.title('Linear Regression using Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
