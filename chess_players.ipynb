```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Comparison_of_top_chess_players_throughout_history')
page
```

```python
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
target_table = soup.find('table', class_ = 'wikitable')
target_table
```

```python
t_head = target_table.find_all('th')
t_head
```

```python
cols = [val.text.strip() for val in t_head]
cols
```

```python
import numpy as np
ind_cells = []
cols = target_table.find_all('td')
for cell in cols:
    individual_cell_value  = cell.text.strip()
    ind_cells.append(individual_cell_value)
data_array = np.array(ind_cells).reshape(-1, 5)
print(data_array)
```

```python
df = pd.DataFrame(data_array , columns= ['Rank', 'Rating', 'Player', 'Date', 'Age'])
df
```

```python
df['Rank'] = df['Rank'].str.replace('(tie)', '').str.strip()
df['Rank']
```

```python
df['Rank'] = [rank[:2] for rank in df['Rank']]
df['Rank']
```

```python
df['Date'] = [date[7:] for date in df['Date']]
df['Date']
```
