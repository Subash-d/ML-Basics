```python
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.365chess.com/top-chess-players.php'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
target_table = soup.find_all('table',class_ = 'table stable')
target_table
```

```python
for i in target_table:
    head = i.find('thead')
    cols = i.find('tbody')
```

```python
head
```

```python
table_col_names = [
    val.text.strip() for val in head.find_all('td')
]
table_col_names
```

```python
table_col_names[0] = 'Number'
```

```python
table_col_names
```

```python
cols
```

```python
table_rows = []
for row in cols.find_all('tr'):
    data = row.find_all('td')
    table_rows.append([val.text.strip() for val in data])
    print([val.text.strip() for val in data])
```

```python
df = pd.DataFrame(table_rows, columns=table_col_names)
```

```python
df.head()
```

```python
df.to_csv('Top200_fide_rated.csv', index=False)
```
